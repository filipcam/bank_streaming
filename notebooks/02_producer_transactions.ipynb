# 02. Producer transactions
Producer wysyła rekordy co 0.5 sekundy do topicu `transactions`

```python
import json, time
from pathlib import Path
from kafka import KafkaProducer

path = Path("../data/transactions.json")
print(path.resolve())

with path.open("r", encoding="utf-8") as f:
    transactions = json.load(f)

producer = KafkaProducer(
    bootstrap_servers="kafka_streaming_lab:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8"),
    request_timeout_ms=5000,
    max_block_ms=5000
)

for tx in transactions:
    producer.send("transactions", key=tx["sender"].encode("utf-8"), value=tx)
    producer.flush()
    print(f"Sent: {tx}")
    time.sleep(0.5)

producer.close()
```
