# 02. Producer transactions
Producer wysyła wiadomości co 0.5 sekundy do topicu `transactions`, analogicznie do laboratoryjnego producenta Kafka [file:11].

```python
from kafka import KafkaProducer
import json, time

with open("../data/transactions.json", "r", encoding="utf-8") as f:
    transactions = json.load(f)

producer = KafkaProducer(
    bootstrap_servers="kafka_streaming_lab:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

for tx in transactions:
    producer.send("transactions", key=tx["sender"].encode("utf-8"), value=tx)
    producer.flush()
    print(f"Sent: {tx}")
    time.sleep(0.5)

producer.close()
```
