# 01. Create Kafka topic
Tworzenie topicu `transactions` przez klienta administracyjnego Kafka

```python
from confluent_kafka.admin import AdminClient, NewTopic

admin = AdminClient({"bootstrap.servers": "kafka_streaming_lab:9092"})
topic_name = "transactions"
new_topic = NewTopic(topic_name, num_partitions=3, replication_factor=1)
fs = admin.create_topics([new_topic])

for topic, f in fs.items():
    try:
        f.result()
        print(f"Topic {topic} created")
    except Exception as e:
        print(f"Topic {topic}: {e}")
```
